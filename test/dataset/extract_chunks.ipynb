```python
import pandas as pd
import glob

# This list will hold all elements from column3 across all CSV files
all_column3_entries = []

# Loop over all CSV files in the current directory (adjust glob pattern if needed)
for filename in glob.glob("datasets_by_category/*.csv"):
    # Read the CSV
    df = pd.read_csv(filename, encoding='utf-8')
    
    # Go through each row in 'column3'
    for cell in df["column 3"]:
        # Ensure the cell is not empty or NaN
        if pd.notna(cell):
            # Split the string by '==='
            parts = cell.split("===")
            # Strip whitespace around each split part
            parts = [p.strip() for p in parts]
            # Add the split parts to our main list
            all_column3_entries.extend(parts)

# Now 'all_column3_entries' is a Python list containing all the split elements
print(all_column3_entries)
```

```python
import pandas as pd
import glob

# This list will hold all chunks across all CSV files
all_chunks = []

# Loop over all CSV files in the datasets_by_category directory
for filename in glob.glob("datasets_by_category/*.csv"):
    try:
        # Read the CSV with semicolons as separators and handle quoting
        df = pd.read_csv(filename, sep=';', encoding='utf-8', quotechar='"')
        
        # Get the column that contains chunks (appears to be the third column)
        # The column name may vary, so we check if 'column 3' exists or use the third column
        if 'column 3' in df.columns:
            chunk_column = 'column 3'
        elif len(df.columns) >= 3:
            chunk_column = df.columns[2]  # Use the third column
        else:
            print(f"Warning: Not enough columns in {filename}")
            continue
        
        # Go through each row in the chunk column
        for cell in df[chunk_column]:
            # Ensure the cell is not empty or NaN
            if pd.notna(cell):
                # Split the string by '==='
                parts = cell.split("===")
                # Strip whitespace around each split part
                parts = [p.strip() for p in parts]
                # Add the split parts to our main list
                all_chunks.extend(parts)
                
        print(f"Processed {filename}: {len(parts)} chunks found")
                
    except Exception as e:
        print(f"Error processing {filename}: {e}")

# Print how many chunks we found
print(f"Total chunks extracted: {len(all_chunks)}")
```
