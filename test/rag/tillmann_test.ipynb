```python
import requests
import time

url_embedding = "http://localhost:8002/embed_text"
url_add = "http://localhost:8003/add"

headers = {
    "accept": "application/json",
    "X-API-Key": "test",
    "Content-Type": "application/json"
}

start = time.time()

to_upload = napkin #transcript
meta = {
    "language": "de",
    "filename": "Napkin Ideas",
}

for i,text in enumerate(to_upload):

    data_embeddeding = {
        "text": text
    }

    try: 
        response = requests.post(url_embedding, headers=headers, json=data_embeddeding)

        embedding = response.json()['embedding']
        #print(embedding)
    except:
        print("Error")
        print(text)

    data_add = {
        "documents": [
            {
                "text": text,
                "embedding": embedding,
                "metadata": {
                    "language": meta["language"],
                    "filename": meta["filename"],
                    "chunk_index": i+1,
                    "total_chunks": len(transcript)
                }
            }
        ]
    }

    response_add = requests.post(url_add, headers=headers, json=data_add)

print("Time: ", time.time()-start)
```

```python
url_query = "http://localhost:8003/query"

headers = {
    "accept": "application/json",
    "X-API-Key": "test",
    "Content-Type": "application/json"
}

data_query = {
    "text": "What should I focus on in a startup?",
    "n_results": 5
}

response = requests.post(url_query, headers=headers, json=data_query)

response_text = ""
for r in response.json()['results']:
    response_text += r['document'] + "\n\n"

print(response_text)
```

```python
import requests

def query_text(text: str, n_results: int = 5) -> str:
    response = requests.post(
        "http://localhost:8003/query",
        headers={
            "accept": "application/json",
            "X-API-Key": "test",
            "Content-Type": "application/json"
        },
        json={"text": text, "n_results": n_results}
    )
    return "\n\n".join(r["document"] for r in response.json()["results"])
```
